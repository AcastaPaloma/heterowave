# Physics sanity checks
Synthetic phantom, parallel-beam sinogram, UBP, and FBP only.

```python
import torch
import matplotlib.pyplot as plt
from heterowave.data.phantoms import PhantomRegion, render_phantom
from heterowave.physics import parallel_beam_project, unfiltered_backprojection, filtered_backprojection
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
speed = render_phantom(64, [
    PhantomRegion('circle', (0.0, 0.0), (0.48, 0.48), 1540.0),
    PhantomRegion('ellipse', (-0.16, 0.10), (0.13, 0.22), 1470.0, 25.0),
    PhantomRegion('rectangle', (0.22, -0.16), (0.11, 0.08), 1580.0, -15.0),
], device=device).unsqueeze(0)
image = (speed - 1500.0) / 100.0
sinogram = parallel_beam_project(image, num_angles=64)
ubp = unfiltered_backprojection(sinogram)
fbp = filtered_backprojection(sinogram)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(14, 3))
for ax, value, title in zip(axes, [image[0, 0], sinogram[0], ubp[0, 0], fbp[0, 0]], ['Phantom', 'Sinogram', 'UBP', 'FBP']):
    ax.imshow(value.detach().cpu(), cmap='gray', aspect='auto')
    ax.set_title(title)
    ax.axis('off')
plt.show()
```
